```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import KFold
import sklearn.cross_validation as cv
import math
```

```python
data = pd.read_csv('data/AC2.csv')
data = data.apply(lambda x: x.fillna(x.median()), axis=0)
data = pd.get_dummies(data, columns=['C{}'.format(x) for x in range(1, 9)])
data.head()
```

```python
def fit_depth(X, Y):
    opt_d = 0
    opt_score = -math.inf
    kf = cv.KFold(len(X), n_folds=10, shuffle=True, random_state=2016)
    for d in range(3, 11):
        rf_classifier = RandomForestClassifier(n_estimators=200, max_depth=d, random_state=2016)
        scores = cv.cross_val_score(rf_classifier, X, Y, cv=kf, scoring='f1')
        score = scores.mean()
        if score > opt_score:
            opt_score, opt_d = score, d
            print("Current f1 measure = [ {} ], depth = [ {} ]".format(opt_score, opt_d))
    return opt_d, opt_score
```

```python
X = data.drop(['Approved'], axis=1)
Y = data['Approved']
d,s = fit_depth(X, Y)
print("Winner depth = {}".format(d))
```
