```python
# just for lulz; experimenting with Central Limit Theorem
import matplotlib
matplotlib.style.use('ggplot')

%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import random
from collections import Counter
N = 300
M = 20000
g = (sum(arr) for arr in ((random.choice([-1, 1]) for i in range(N)) for j in range(M)))
df = pd.DataFrame(g)
df.hist(bins=15)
```
